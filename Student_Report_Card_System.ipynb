```python
name=str(input("Name : "))
school=str(input("School Name : "))
java=int(input("Enter Java Marks out of 100 : "))
c=int(input("Enter C++ Marks out of 100 : "))
jsp=int(input("Enter JSP Marks out of 100 : "))
python=int(input("Enter Python Marks out of 100 : "))
javascript=int(input("Enter JavaScript Marks out of 100 : "))
go=int(input("Enter Go Marks out of 100 : "))

print("------------------------------")
print("School Name : " ,school)
print("Student Name : " ,name)
print("------------------------------")

#JAVA
if java>=80 and java<=100:
  print("Java       = ",java,"/100 - Grade A")
elif java<=80 and java>=40:
  print("Java       = ",java,"/100 - Grade B")
elif java<=40 and java>=0:
  print("Java       = ",java,"/100 - Grade C")
else:
  print("Please Enter Valid Number")

#c++
if c>=80 and c<=100:
  print("C++        = ",c,"/100 - Grade A")
elif c>=40 and c<=80:
  print("C++        = ",java,"/100 - Grade B")
elif c<=40 and c>=0:
  print("C++        = ",c,"/100 - Grade C")
else:
  print("Please Enter Valid Number")

#jsp
if jsp>=80 and jsp<=100:
  print("JSP        = ",jsp,"/100 - Grade A")
elif jsp>=40 and jsp<=80:
  print("JSP        = ",java,"/100 - Grade B")
elif jsp<=40 and jsp>=0:
  print("JSP        = ",jsp,"/100 - Grade C")
else:
  print("Please Enter Valid Number")

#python
if python>=80 and python<=100:
  print("Python     = ",python,"/100 - Grade A")
elif python>=40 and python<=80:
  print("Python     = ",python,"/100 - Grade B")
elif python<=40 and python>=0:
  print("Python     = ",python,"/100 - Grade C")
else:
  print("Please Enter Valid Number")

#javascript
if javascript>=80 and javascript<=100:
  print("Javascript = ",javascript,"/100 - Grade A")
elif javascript>=40 and javascript<=80:
  print("Javascript = ",javascript,"/100 - Grade B")
elif javascript<=40 and javascript>=0:
  print("Javascript = ",javascript,"/100 - Grade C")
else:
  print("Please Enter Valid Number")

#go
if go>=80 and go<=100:
  print("Go         = ",go,"/100 - Grade A")
elif go>=40 and go<=80:
  print("Go         = ",go,"/100 - Grade B")
elif go<=40 and go>=0:
  print("Go         = ",go,"/100 - Grade C")
else:
  print("Please Enter Valid Number")

print("------------------------------")

#total
total = java+c+jsp+python+javascript+go
print("Total = " ,total)

#percentage
percentage=total/600*100
print("Percentage = ",percentage,"%")

#overall
if percentage>=80:
  print("Orerall Grade = A")
elif percentage>=40 and percentage<=80:
  print("Overall Grade = B")
elif percentage>=35 and percentage<=40:
  print("Overall Grade = C")
else:
  print("You are Fail")
```
